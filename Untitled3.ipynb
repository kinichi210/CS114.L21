```python
# số lần submit
n_submit = df1.groupby('user_id').size()

# so problem tham gia giai
Participate = df1.groupby('user_id')['problem_id'].nunique()

# so problem xanh
solved_problem = df1[(df1['is_final'] == 1) & (df1['pre_score'] == 10000)].groupby('user_id')['problem_id'].nunique()

# tong pre_score co is_final = 1
pre_score = df1[df1['is_final'] == 1].groupby('user_id')['pre_score'].sum()

data = pd.DataFrame({'n_submits': n_submit,
                     'Participate': Participate,
                     'solved_problems': solved_problem,
                     'pre_score: Is_final = 1':pre_score})
# Tỷ lệ hoàn thành Problem đã tham gia
data['solved_problem/Participate'] = data['solved_problems']/data['Participate']
# Tỷ lệ Problem tham gia giải/số submittions
data['Participate/n_submit'] = data['Participate']/data['n_submits']
data.reset_index(inplace=True)
```
